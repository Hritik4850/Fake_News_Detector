```python
import pandas as pd
true_df = pd.read_csv(r"C:\Users\Sunshine\Downloads\Fake.csv")
fake_df = pd.read_csv(r"C:\Users\Sunshine\Downloads\True.csv")

true_df['label'] = 0
fake_df['label'] = 1

true_df.head()
#print(true_df)

fake_df.head()
#print(fake_df)

true_df = true_df[['text','label']]                          #we drop the 'text' and 'label'
fake_df = fake_df[['text','label']]                          #we drop the 'text' and 'label'

dataset = pd.concat([true_df , fake_df])                     #to merge two dataframes
 
dataset['label'].value_counts()                              #to verify the dataset as a balanced distribution

dataset = dataset.sample(frac = 1)                           #As we have concatenated the dataframes, 
                                                             #we should shuffle the dataset. 
                                                             #Else, our model learns in order and results in worse accuracy.

dataset.head(20)                                            #check the output
```
